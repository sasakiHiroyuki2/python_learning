```python
fig,ax = plt.subplots()
x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ["spam","ham","egg"]
width = 0.4
ax.bar(x,y1,width = width,tick_label=labels,label = "y1")

x2 = [num + width for num in x]
ax.bar(x2,y2,width = width,tick_label=labels,label = "y2")
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots()
y_total = [num1 + num2 for num1,num2 in zip(y1,y2)]
ax.bar(x,y_total,tick_label=labels,label = "y1")
ax.bar(x,y2,tick_label=labels,label = "y2")
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots()
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)
ax.scatter(x[0:10],y[0:10],marker = "v",label="triangle down", c="red", s=30)
ax.scatter(x[10:20],y[10:20],marker = "^",label="triangle up", c="green", s=50)
ax.scatter(x[20:30],y[20:30],marker = "s",label="squea", c="yellow",s=60 ,alpha=0.5)
ax.scatter(x[30:40],y[30:40],marker = "*",label="star")
ax.scatter(x[40:50],y[40:50],marker = "x",label="X")
ax.legend()
plt.show()
```

```python
np.random.seed(123)
x = np.random.randn(1000)
fig,ax = plt.subplots()
counts,edges,patches = ax.hist(x,bins = 25,edgecolor="red" ,color="red")

x_fit = (edges[:-1] + edges[1:])/2
y = 1000* np.diff(edges) * np.exp(-x_fit ** 2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit,y,color="#ff0000",linewidth=4,linestyle="--")

ax.set_xlabel("xlabel",family="fantasy",size=20,weight="bold")
ax.set_ylabel("ylabel",family="cursive",size=40,weight="light")
ax.set_title("title",family="monospace",size=25,weight="heavy")
plt.show()
```

```python
fontdict = {
    "family":"fantasy",
    "size":20,
    "weight":"normal"
}
fig,ax = plt.subplots()

ax.set_xlabel("xlabel",fontdict = fontdict)
ax.set_ylabel("ylabel",fontdict = fontdict)

ax.set_title("title",fontdict = fontdict,size=40)
plt.show()
```
